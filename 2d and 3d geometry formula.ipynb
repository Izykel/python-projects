```python
import math as mth
while True:
    print("select: 1 for 2D geometry")
    print("select: 2 for 3D geometry")
    choice = int(input())

    if choice == 1 and choice !=0 and choice < 2:
        print("pick a 2D shape")
        print('choose 1 for square:>_')
        print('choose 2 for rectangle :>_')
        print('choose 3 for triangle:>_')
        print('choose 4 for equilateral triangle:>_')
        print('choose 5 for parallelogram:>_')
        print('choose 6 for trapezoid:>_')
        print('choose 7 for circle:>_')
        print('choose 8 for sector of circle:>_')
        print('choose 9 for ellipse:>_')
        print('choose 10 for annulus:>_')

        choice = int(input())

        if choice == 1 and choice != 0:
            side = float(input('please enter the side value of the square:>_\n'))
            area = side ** 2
            perimeter = 4 * side
            if side !=0:
                print('shape:square')
                print(f'area:{area}')
                print(f'circumference:{perimeter}')
            else:
                print('check your side value and ry again')

        elif choice == 2 and choice != 0:
            length = float(input('please enter the length of the rectangle:>_\n'))
            width = float(input('please enter the width of the rectangle:>_\n'))
            area = length * width
            perimeter = 2 * length + 2 * width
            if length !=0 and width !=0:
                print('shape:rectangle')
                print(f'area:{area}')
                print(f'circumference:{perimeter}')
            else:
                print('check your values and try again')

        elif choice == 3 and choice != 0:
            base = float(input('please enter the base value of the triangle:>_\n'))
            height = float(input('please enter the height value of the triangle:>_\n'))
            a = float(input('please enter the a value:>_\n'))
            b = float(input('please enter the b value:>_\n'))
            c = float(input('please enter the c value:>_\n'))
            area = 0.5 * base * height
            perimeter = a + b + c
            if base !=0 and height !=0 and a !=0 and b != 0 and c:
                print('shape:triangle')
                print(f'area:{area}')
                print(f'perimeter:{perimeter}')
            else:
                print('check your values and try again')

        elif choice == 4 and choice != 0:
            side = float(input('please enter the side value of the equilateral triangle:>_\n'))
            area = (mth.sqrt(3)/4) * side ** 2
            height = (mth.sqrt(3)/2) * side
            if side !=0 and area != 0 and height != 0:
                print('shape:equilateral triangle')
                print(f'area:{area}')
                print(f'circumference:{height}')
            else:
                print('check your values and try again')

        elif choice == 5 and choice != 0:
            base = float(input('please enter the radius of the circle:>_\n'))
            height = float(input('please enter the diameter of the circle:>_\n'))
            side = float(input('please enter the diameter of the circle:>_\n'))
            area = base * height
            perimeter = 2 * side + 2 * height
            if base !=0 and height !=0 and side !=0 and area !=0:
                print('shape:parallelogram')
                print(f'area:{area}')
                print(f'perimeter:{perimeter}')
            else:
                print('check your values and try again')

        elif choice == 6 and choice != 0:
            top_base = float(input('please enter the top base value:>_\n'))
            bottom_base = float(input('please enter the bottom base value:>_\n'))
            left_side = float(input('please enter the left side value:>_\n'))
            right_side = float(input('please enter the right side value:>_\n'))
            height = float(input('please enter the height value:>_\n'))
            area = 0.5 * (top_base + bottom_base) * height
            perimeter = top_base + bottom_base + left_side + right_side
            if top_base !=0 and bottom_base !=0 and left_side !=0 and right_side and height !=0:
                print('shape:trapezoid')
                print(f'area:{area}')
                print(f'perimeter:{perimeter}')
            else:
                print('check your values and try again')

        elif choice == 7 and choice != 0:
            Radius = float(input('please enter the radius of the circle:>_\n'))
            Diameter = float(input('please enter the diameter of the circle:>_\n'))
            area = mth.pi * Radius ** 2
            circumference = mth.pi * Diameter
            if Radius !=0 and Diameter !=0:
                print('shape:circle')
                print(f'area:{area}')
                print(f'circumference:{circumference}')
            else:
                print('check your values and try again')

        elif choice == 8 and choice != 0:
            Radius = float(input('please enter the radius of the circle:>_\n'))
            theta = float(input('please enter the angle subtended by of the circle:>_\n'))
            area = 0.5 * theta * Radius ** 2
            arc_length = theta * Radius
            if Radius !=0 and theta !=0:
                print('shape:sector of circle')
                print(f'area:{area}')
                print(f'arc length:{arc_length}')
            else:
                print('check your values and try again')

        elif choice == 9 and choice != 0:
            a = float(input('please enter the semi major a of the ellipse:>_\n'))
            b = float(input('please enter the semi minor of the ellipse:>_\n'))
            area = mth.pi * a * b
            perimeter = mth.pi * (3 * (a + b) - mth.sqrt((a + 3*b) * (b + 3*a)))
            if  a !=0 and b !=0:
                print('shape:ellipse')
                print(f'area:{area}')
                print(f'perimeter:{perimeter}')
            else:
                print('check your values and try again')

        elif choice == 10 and choice != 0:
            inner_radius = float(input('please enter the inner radius of the annulus:>_\n'))
            outer_radius = float(input('please enter the outer radius of the annulus:>_\n'))
            average_radius = 0.5 * (inner_radius + outer_radius)
            width = outer_radius - inner_radius
            area = mth.pi * (outer_radius **2 - inner_radius **2)
            if inner_radius !=0 and outer_radius !=0:
                print('shape:annulus')
                print(f'area:{area}')
                print(f'width:{width}')
                print(f'average radius:{average_radius}')
            else:
                 print('check your values and try again')

        else:
            print('please,pick a valid shape')



    elif choice == 2 and choice !=0:
        print("pick a 3D shape")
        print('choose 1 for cube:>_')
        print('choose 2 for rectangular solid:>_')
        print('choose 3 for sphere:>_')
        print('choose 4 for right circular cylinder:>_')
        print('choose 5 for torus:>_')
        print('choose 6 for pyramid:>_')
        print('choose 7 for right circular cone:>_')
        print('choose 8 for frustum of a cone:>_')
        print('choose 9 for  square pyramid:>_')
        print('choose 10 for regular tetrahedron:>_')

        choice = int(input())

        if choice == 1 and choice != 0:
            side = float(input('please enter the sides value of the cube:>_\n'))
            surface_area = 6 * side ** 2
            volume = side ** 3
            if side !=0:
                print('shape:cube')
                print(f'surface area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        elif choice == 2 and choice != 0:
            length = float(input('please enter the length of the rectangular solid:>_\n'))
            width = float(input('please enter the width of the rectangular solid:>_\n'))
            height = float(input('please enter the height of the rectangular solid:>_\n'))
            surface_area = 2 * length * width + 2 * length * height + 2 *  width * height
            volume = length * width * height
            if length !=0 and width !=0 and height !=0:
                print('shape:rectangular solid')
                print(f'surface area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        elif choice == 3 and choice != 0:
            Radius = float(input('please enter the radius of the sphere:>_\n'))
            surface_area = 4 * mth.pi * Radius ** 2
            volume = 1.33 * mth.pi * Radius ** 3
            if Radius!=0:
                print('shape:sphere')
                print(f'area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        elif choice == 4 and choice != 0:
            Radius = float(input('please enter the radius of the circular cylinder:>_\n'))
            height = float(input('please enter the height of the circular cylinder:>_\n'))
            surface_area = 2 * mth.pi * Radius * height + 2 * mth.pi * Radius ** 2
            volume = mth.pi * Radius ** 2 * height
            if Radius !=0 and height !=0:
                print('shape:right circular cylinder')
                print(f'surface_area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        elif choice == 5 and choice != 0:
            tube_radius = float(input('please enter the  tube radius of the torus:>_\n'))
            torus_radius = float(input('please enter the torus_radius of the torus:>_\n'))
            surface_area = 4 * mth.pi ** 2 * tube_radius * torus_radius
            volume = 2 * mth.pi ** 2 * tube_radius ** 2 * torus_radius
            if tube_radius !=0 and torus_radius !=0:
                print('shape:Torus')
                print(f'surface area:{surface_area}')
                print(f'volume:{volume}')
            else:
               print ('check your values and try again')

        elif choice == 6 and choice != 0:
            area_of_base = float(input('please enter the area of base of the cone:>_\n'))
            height = float(input('please enter the height of the cone:>_\n'))
            volume = 0.33 * area_of_base * height
            if area_of_base !=0 and height !=0:
                print('shape:General cone')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')


        elif choice == 7 and choice != 0:
            Radius = float(input('please enter the radius of the circular cone:>_\n'))
            height = float(input('please enter the height of the circular cone:>_\n'))
            surface_area = mth.pi * Radius * (mth.sqrt(Radius ** 2 + height ** 2) + mth.pi * Radius ** 2)
            volume = 0.33 * mth.pi * Radius ** 2 * height
            if Radius !=0 and height !=0:
                print('shape:Right circular cone')
                print(f'surface area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        elif choice == 8 and choice != 0:
            top_radius = float(input('please enter the top radius of the frustum:>_\n'))
            base_radius = float(input('please enter the base radius of the frustum:>_\n'))
            height = float(input('please enter the top height of the frustum:>_\n'))
            slant_height = float(input('please enter the slant height of the frustum:>_\n'))
            volume= 1.0467 * (top_radius ** 2 + top_radius * base_radius + base_radius) * height
            surface_area = mth.pi * slant_height * ( base_radius + top_radius) + mth.pi * top_radius ** 2 + mth.pi * base_radius ** 2
            if top_radius !=0 and base_radius !=0 and height !=0 and slant_height !=0:
                print('shape:Frustum of a cone')
                print(f'surface area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        elif choice == 9 and choice != 0:
            height = float(input('please enter the height of the pyramid:>_\n'))
            side = float(input('please enter the side value of the pyramid:>_\n'))
            surface_area = side * (side + mth.sqrt(side ** 2 + 4 * height ** 2))
            volume = 0.33 * side ** 2 * height
            if height !=0 and side !=0:
                print('shape:Square pyramid')
                print(f'area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        elif choice == 10 and choice != 0:
            side = float(input('please enter the side value of the tetrahedron:>_\n'))
            volume = 0.083 * mth.sqrt(2 * side ** 3)
            surface_area = mth.sqrt(3 * side ** 2)
            if side !=0:
                print('shape:Regular Tetrahedron')
                print(f'surface area:{surface_area}')
                print(f'volume:{volume}')
            else:
                print('check your values and try again')

        else:
            print('please,pick a valid 3D shape')

    else:
        print("pick a valid choice")
        break
```
